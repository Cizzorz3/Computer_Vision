```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from PIL import Image
```

```python
## to open an image
pic = Image.open("puppy-care-guide-for-new-parents.jpg");
```

```python
##to open the image
pic
```

```python
type(pic)
```

```python
##to convert the type to an array
pic_arr=np.asarray(pic)
```

```python
type(pic_arr)
```

```python
## to see the shape of the arr
pic_arr.shape
```

```python
## to show the image in the array
plt.imshow(pic_arr)
```

```python
##to access specific channel
pic_red = pic_arr.copy()
```

```python
pic_red.shape
```

```python
## RGB
#Red channel values 0 no red pure black 255 means full pure red
#in grey scale more white means  more red
#cmap color map gray scale image
plt.imshow(pic_red[:,:,0],cmap='gray')
```

```python
#green channel
plt.imshow(pic_red[:,:,1],cmap='gray')
```

```python
# blue channel
plt.imshow(pic_red[:,:,2],cmap='gray')
```
