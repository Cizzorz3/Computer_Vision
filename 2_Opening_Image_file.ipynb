```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
## to read an image 
##Note in case of wrong path it won't show an error it will only return non type file
img=cv2.imread("puppy-care-guide-for-new-parents.jpg")
```

```python
## the type is automatically an array you dont have to convert it
type(img)
```

```python
img.shape
```

```python
## Note matplotlib expect R G B
## OpenCV reads as B G R
plt.imshow(img)
```

```python
## to change the color format
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
## to read image in specific format
img_gray=cv2.imread('puppy-care-guide-for-new-parents.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
## Note we read it as greyscale so it only has 2 channels
img_gray.shape
```

```python
## note you have to put cmap=gray in order to view it in greyscale
plt.imshow(img_gray,cmap='gray')
```

```python
##to resize img
#Note 1000 is crossponding to 2048
#And 600 is crossponding to 1365
new_img=cv2.resize(fix_img,(1000,600))
```

```python
plt.imshow(new_img)
```

```python
w_ratio=0.8
h_ratio=0.5
```

```python
## to resize by a ratio 
new_img=cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
##to flip images 
## 0 ,1 ,-1
new_img=cv2.flip(fix_img,0)
```

```python
plt.imshow(new_img)
```

```python
type(fix_img)
```

```python
## to save an image from cv2
cv2.imwrite('new_photo.jpg',fix_img)
```
